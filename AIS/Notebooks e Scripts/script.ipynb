```python
import pandas as pd

gdp = pd.read_csv("DataSetOriginal/countries_gdp_hist.csv", delimiter = ';')
inflation = pd.read_excel('DataSetProfs/Inflation-data.xlsx', sheet_name= 'hcpi_a')
govexp = pd.read_csv("DataSetOriginal/goverment_expenditure.csv", delimiter = ';')
```

```python
gdp = gdp.pivot(index='country_code', columns='year', values='total_gdp')
```

```python
inflation.head()
```

```python
inflation = inflation.drop(['IMF Country Code', 'Country', 'Indicator Type', 'Series Name', 'Note'],axis=1)
```

```python
inflation.isna().sum()
```

```python
# Separate the first column containing the country code
country_code = inflation.iloc[:, 0]

# Convert the remaining data to a numeric data type
inflation = inflation.iloc[:, 1:].apply(pd.to_numeric)

# Compute the mean of each row
means = inflation.mean(axis=1)

# Replace the NaN values in each row with the mean value
for i in range(len(inflation)):
    inflation.iloc[i] = inflation.iloc[i].fillna(means[i])

# Add the first column back to the dataset
inflation = pd.concat([country_code, inflation], axis=1)
```

```python
govexp = govexp.drop(['Country Name', 'Indicator Name', 'Indicator Code', 'Unnamed: 66'],axis=1)
```

```python
govexp = govexp.dropna(axis=1, how='all')
```

```python
govexp.head()
```

```python
govexp = govexp.drop(['1960'], axis=1)

# Separate the first column containing the country code
country_code = govexp.iloc[:, 0]

# Convert the remaining data to a numeric data type
govexp = govexp.iloc[:, 1:].apply(pd.to_numeric)

# Compute the mean of each row
means = govexp.mean(axis=1)

# Replace the NaN values in each row with the mean value
for i in range(len(govexp)):
    govexp.iloc[i] = govexp.iloc[i].fillna(means[i])

# Add the first column back to the dataset
govexp = pd.concat([country_code, govexp], axis=1)
```

```python
govexp = govexp.loc[govexp['Country Code'].str.len() <= 3]
```

```python
govexp.to_csv('Govermnent_expenditure.csv', index=False)
```

```python
inflation = inflation.loc[inflation['Country Code'].str.len() <= 3]
```

```python
inflation.to_csv('Inflation.csv', index=False)
```

```python
gdp = gdp.drop(['1960'], axis=1, errors='ignore')
# Separate the first column containing the country code
country_code = gdp.iloc[:, 0]

# Convert the remaining data to a numeric data type
gdp = gdp.iloc[:, 1:].apply(pd.to_numeric)

# Compute the mean of each row
means = gdp.mean(axis=1)

# Replace the 0 values in each row with the mean value
for i in range(len(gdp)):
    row = gdp.iloc[i]
    if 0 in row.values:
        row[row == 0] = means[i]

# Add the first column back to the dataset
gdp = pd.concat([country_code, gdp], axis=1)
```

```python
gdp.head()
```

```python
gdp.to_csv('Total_GDP.csv', index=True)
```

```python
gdp.reset_index(inplace=True)
melted_gdp = pd.melt(gdp, id_vars=['country_code'], var_name='year', value_name='gdp')
```

```python
melted_gdp.to_csv('Total_GDP.csv', index=False)
```

```python
import pandas as pd

literacy = pd.read_csv("DataSetOriginal/API_SE.ADT.LITR.ZS_DS2_en_csv_v2_5356271.csv", delimiter = ',')
```

```python
literacy.head()
```

```python
literacy = literacy.drop(['Country Name','Indicator Name', 'Indicator Code'],axis=1)
```

```python
literacy = literacy.drop('Unnamed: 66', axis=1)
```

```python
literacy.head()
```

```python
literacy.isna().sum()
```

```python
# Separate the first column containing the country code
country_code = literacy.iloc[:, 0]

# Convert the remaining data to a numeric data type
literacy = literacy.iloc[:, 1:].apply(pd.to_numeric)

# Compute the mean of each row
means = literacy.mean(axis=1)

# Replace the NaN values in each row with the mean value
for i in range(len(literacy)):
    literacy.iloc[i] = literacy.iloc[i].fillna(means[i])

# Add the first column back to the dataset
literacy = pd.concat([country_code, literacy], axis=1)
```

```python
literacy.head()
```

```python
literacy.dropna(inplace=True)
```

```python
literacy.isna().sum()
```

```python
literacy.head()
```

```python
literacy.to_csv('Literacy Rate.csv', index=True)
```

```python
melted_literacy = pd.melt(literacy, id_vars=['Country Code'], var_name='year', value_name='literacy_rate')
```

```python
melted_literacy.to_csv('Literacy Rate.csv', index=False)
```
